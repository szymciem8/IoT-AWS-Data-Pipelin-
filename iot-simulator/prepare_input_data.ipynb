```python
import pandas as pd
```

```python
df = pd.read_csv('data/predictive_maintenance_dataset.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
df.tail()
```

```python
distict_devices = df['device'].unique()
all_dattes = df['date'].unique()
number_of_days = len(all_dattes)
sorted_dates = sorted(all_dattes)
start_date = sorted_dates[0]
end_date = sorted_dates[-1]

# Format and print all info
print("Number of rows: {}".format(len(df)))
print("Number of devices: {}".format(len(distict_devices)))
print("Start date: {}".format(start_date))
print("End date: {}".format(end_date))
print("Number of days: {}".format(number_of_days))
```

```python
# Create seperate dfs for each device
device_dfs = {}
for device in distict_devices:
    device_dfs[device] = df[df['device'] == device]
    # Sort by date
    device_dfs[device].sort_values(by=['date'], inplace=True)
    # Rester index
    device_dfs[device].reset_index(inplace=True, drop=True)
```

```python
device_dfs['S1F01E6Y'].head()
```

```python
# Calculate the average number of days each device
avg_days_per_device = 0
for device in device_dfs:
    avg_days_per_device += len(device_dfs[device]['date'].unique())
avg_days_per_device = avg_days_per_device / len(device_dfs)
avg_days_per_device
```

```python
# Save data for each device in a xlxs file in a seperate sheet
with pd.ExcelWriter('data/predictive_maintenance_dataset.xlsx') as writer:
    for device in device_dfs:
        device_dfs[device].to_excel(writer, sheet_name=device, index=False)
```

```python
# Read data from the excel file and create a dataframe for each device
device_dfs = {}
with pd.ExcelFile('data/predictive_maintenance_dataset.xlsx') as xls:
    # Get distinct all sheet names
    distict_devices = xls.sheet_names
    for device in distict_devices:
        device_dfs[device] = pd.read_excel(xls, sheet_name=device)
```
